```python
def hano(n,a,b,c):
    '''
    汉诺塔的递归实现
    '''
    if n == 1:
        print(a,"-->",c)
        return None
    if n == 2:
        print(a,"-->",b)
        print(a,"-->",c)
        print(b,"-->",c)
        return None
    hano(n-1,a,c,b)
    print(a,"-->",c)
    hano(n-1,b,a,c)
a = "A"
b = "B"
c = "C"
n = 9
hano(n,a,b,c)
```
