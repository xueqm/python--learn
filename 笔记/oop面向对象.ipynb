```python
class Fish():
    def __init__(self,name):
        self.name = name
    def swim(self):
        print("I am swimming")
class Bird():
    def __init__(self,name):
        self.name = name
    def fly(self):
        print("I am flying")
class Person():
    def __init__(self,name):
        self.name = name
    def work(self):
        print("I am working")   
# 多继承的例子
class SuperMan(Person,Bird,Fish):
    def __init__(self,name):
        self.name = name

s = SuperMan("yueyue")
s.fly()
s.swim()
s.work()
# 单继承的例子
```

```python
class Student():
    def __init__(self,name,age):
        self.name = name
        self.age = age
        self.setName(name)
    def intro(self):
        print("Hai,my name is {0}".format(self.name))
    def setName(self,name):
        self.name = name.upper()
s1 = Student("LIU Ying",92)
s2 = Student("michi stangle",24)

s1.intro()
s2.intro()
```

```python
class Person():
    '''
    这是一个人，一个高尚的人，一个脱离了低级趣味的人（文档）
    '''
    def fget(self):
        return self._name*2
    def fset(self,name):
        self._name = name.upper()
    def fdel(self):
        self._name = "NoName"
    name = property(fget,fset,fdel,"对name进行下操作啦")
```

```python
p1 = Person()
p1.name = "TuLing"
print(p1.name)
```
